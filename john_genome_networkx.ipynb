```python
nxpdf = nx.from_pandas_edgelist(pdf, source='Kegg_Pathway_Number', target='KAMIAK_Code')
```

```python
nxodf = nx.from_pandas_edgelist(odf, source=0, target=1)
```

```python
t_graph = nx.compose(nxpdf, nxodf)
```

```python
triangles = nx.triangles(t_graph)
```
